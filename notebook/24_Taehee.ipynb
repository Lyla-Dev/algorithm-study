#### 문자 반복 출력하기

#### 제출 답안

````python
def solution(my_string, n):
    answer = ''
    for c in my_string:
        answer += c * n
    return answer
````

````python
print(solution('hello', 10))
````

#### 다른 풀이

**join() 함수**

- 문자열 메서드
- 문자열들을 하나의 문자열로 이어 붙일 때 사용

**예시**
```python
','.join(['a', 'b', 'c'])
```
결과: `'a,b,c'`


```python
''.join(['a', 'b', 'c'])
```
결과: `'abc'`

````python
def solution2(my_string, n):
    return ''.join(c * n for c in my_string)
````

````python
print(solution2('hello', 10))
````
