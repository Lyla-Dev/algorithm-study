### 짝수 홀수 개수

#### 제출 답안

```python
def solution(num_list):
    count_even = 0
    for num in num_list:
        if num % 2 == 0:
            count_even += 1

    return [count_even, len(num_list)-count_even]
```

```python
print(solution([1,2,3,4,5]))
print(solution([1,3,5,7]))
```

#### 다른 풀이

```python
# 천재 같다
def solution2(num_list):
    answer = [0,0]
    for num in num_list:
        answer[num % 2] += 1
    return answer
```

```python
print(solution2([1,2,3,4,5]))
print(solution2([1,3,5,7]))
```
